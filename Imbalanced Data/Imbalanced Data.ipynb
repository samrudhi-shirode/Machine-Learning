# Data Gathering

# Perform data quality checks

```python
df.shape
```

```python
df.info()
```

```python
# check for duplicated rows
df.duplicated().sum()
```

```python
df = df.drop_duplicates(keep='first')
```

```python
# check for missing values
df.isna().sum()
```

# Separate X and Y features

```python
df.columns
```

```python
# remove insignificant columns in X
X = df.drop(columns=['id','loan_status'])
Y = df[['loan_status']]
```

```python
X.head()
```

```python
Y.head()
```

# Data Cleaning and Data Preprocessing

```python
from sklearn.pipeline import make_pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler,OneHotEncoder
from sklearn.compose import ColumnTransformer

cat = list(X.select_dtypes(include='str').columns)
con = list(X.select_dtypes(include='number').columns)

num_pipe = make_pipeline(
    SimpleImputer(strategy='median'),
    StandardScaler()
)

cat_pipe = make_pipeline(
    SimpleImputer(strategy='most_frequent'),
    OneHotEncoder(handle_unknown='ignore',sparse_output=False)
)

pre = ColumnTransformer([
    ('cat',cat_pipe,cat),
    ('con',num_pipe,con)
]).set_output(transform='pandas')

X_pre = pre.fit_transform(X)
X_pre.head()
```

# Separate the data into training and testing
stratify=Y

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(X_pre,Y,train_size=0.66,stratify=Y,random_state=21)
print(xtrain.shape)
print(xtest.shape)
print(ytrain.shape)
print(ytest.shape)
```

# Model Building

```python
from sklearn.linear_model import LogisticRegression
model1 = LogisticRegression()
model1.fit(xtrain,ytrain)
```

```python
from sklearn.metrics import classification_report
ypreds_base = model1.predict(xtest)
print(classification_report(ytest,ypreds_base))
```

```python
model2 = LogisticRegression(class_weight='balanced') # imbalanced scenario
# class_weight='balanced' will add penalty to the minority classes.
model2.fit(xtrain,ytrain)
```

```python
ypreds_balanced = model2.predict(xtest)
print(classification_report(ytest,ypreds_balanced))
```

    SMOTE : Synthetic Minority Over Sampling Technique

# Restart your kernel and run all the cells

```python
from imblearn.over_sampling import SMOTE,ADASYN

smote = SMOTE()
x_sampl1,y_sampl1 = smote.fit_resample(xtrain,ytrain)
```

```python
y_sampl1.value_counts()
```

```python
model3 = LogisticRegression()
model3.fit(x_sampl1,y_sampl1)
```

```python
ypreds_smote = model3.predict(xtest)
print(classification_report(ytest,ypreds_smote))
```

# ADASYN

```python
ada = ADASYN()
x_samp2, y_samp2 = ada.fit_resample(xtrain,ytrain)
```

```python
y_samp2.value_counts()
```

```python
model4 = LogisticRegression()
model4.fit(x_samp2,y_samp2)
```
