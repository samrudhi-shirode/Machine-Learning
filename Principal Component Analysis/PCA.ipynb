# Separate X and Y features

```python
X = df.drop(columns='Weight')
Y = df[['Weight']]
```

```python
from Data_Cleaning_and_Preprocessing.Data_Cl_Pre import data_preparation_cat_con
```

```python
X_pre = data_preparation_cat_con(X)
X_pre.head()
```

# Separate training and testing features

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(X_pre,Y,train_size=0.75,random_state=21)
```

# Create the principal components using PCA

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca.fit(X_pre)
```

```python
X_pca = pca.transform(X_pre)
X_pca
```

# Create new data using the principal components

```python
X_new = pd.DataFrame(X_pca)
X_new.head()
```

```python
X_new.columns = ['PC1','PC2']
X_new.head()
```

```python
x_newtrain,x_newtest,y_train,y_test = train_test_split(X_new,Y,train_size=0.8,random_state=21)
```

# Model Building

```python
from sklearn.ensemble import RandomForestRegressor
rn = RandomForestRegressor(n_estimators=50,max_depth=5)
rn.fit(x_newtrain,y_train)
```

```python
rn.score(x_newtrain,y_train)
```

```python
rn.score(x_newtest,y_test)
```
